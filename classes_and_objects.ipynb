# Classes and Objects in Python

_Object-Oriented Programming_ (OOP) is one of the most important topics in programming. Python is no exception and provides a simple way to create **classes** and **objects**. This notebook is NOT meant to be a complete introduction to OOP. Instead, it serves as a quick reminder of the basic concepts, showing how classes are implemented in Python and how objects are created and used.

### What is a Class?

A _class_ is a blueprint or template for creating objects. It defines the structure and behavior that its objects will have. Think of a class as a cookie cutter and objects as the cookies cut from that template.

In Python, you can create classes using the `class` keyword.

**Creating classes**

When you create a class, you specify the `attributes`(data) and `methods` (functions) that objects of that class will have. Attributes are defined as variables within the class, and methods are defined as functions.

For example, you can design a "Car" class with attributes such as "color" and "speed," along with methods like "accelerate."

### What is an Object?

An _object_ is a fundamental unit in Python that represents a real-world entity or concept.

Every object has two main characteristics:

**State**

The attributes or data that describe the object. For your "Car" object, this might include attributes like "color", "speed", and "fuel level".

**Behavior**

The actions or methods that the object can perform. In Python, methods are functions that belong to objects and can change the object's state or perform specific operations.

**Instantiating objects**

Once you've defined a class, you can create individual objects (instances) based on that class. Each object is independent and has its own set of attributes and methods.

To create an object, you use the class name followed by parentheses, so: `my_car = Car()`

**Interacting with objects**

You interact with objects by calling their methods or accessing their attributes using dot notation `.`

For example, if you have a Car object named `my_car`, you can set its color with `my_car.color = "blue"` and accelerate it with `my_car.accelerate()` if there's an accelerate method defined in the class.

## Class `Circle`

As an example, let's create a class called `Circle`. To define a class in Python, we use the `class` keyword followed by the class name. We then end the line with a colon (`:`) and write the class body below it using indentation.




````python
class Circle:
    pass
````

In older Python code, you may also see `(object)` written after the class name:

```python
class Circle(object):
    pass
```

However, in Python 3, explicitly inheriting from `object` is unnecessary because all classes already do so by default.

## Constructor

The body of a class almost always begins with the **constructor**. The constructor is a special method that initializes the object's **data attributes** when a new instance of the class is created.

In Python, the constructor has the following form:

```python
def __init__(self, attributes):
    # Initialize the object's attributes
```

* `__init__` is a special method that is automatically called whenever a new object is created.
* `self` refers to the current instance of the class, allowing you to access and modify its attributes and methods.
* `attributes` represents any additional values required to initialize the object.


For our example, to define a `Circle` object, we need to specify its `radius` and its `colour`. 

````python
class Circle:
    def __init__(self, radius, color):
        self.radius = radius
        self.color = color
````

## Instance methods

Besides the constructor, a class can define **methods**, which are functions that describe the behavior of its objects.

For example, the `add_radius()` method increases the radius of a `Circle` object by a given amount:

```python
def add_radius(self, r):
    self.radius = self.radius + r
    return self.radius
```

Here's what each line does:

- `self` refers to the current `Circle` object.
- `r` is the amount that will be added to the radius.
- `self.radius = self.radius + r` updates the object's `radius` attribute by adding `r` to its current value.
- `return self.radius` returns the updated radius.

So

````python
class Circle:
    def __init__(self, radius, color):
        self.radius = radius
        self.color = color
    
    def add_radius(self, r):
        self.radius = self.radius + r
        return self.radius
````

We can add another method to display the circle. To do this, we include the `drawCircle` method.

````python
from matplotlib import pyplot as plt

class Circle(object):
    
    # Constructor
    def __init__(self, radius=3, color='blue'): # We can initialize the attributes with default values
        self.radius = radius
        self.color = color 
    
    # Method
    def add_radius(self, r):
        self.radius = self.radius + r
        return(self.radius)
    
    # Method
    def drawCircle(self):
        plt.gca().add_patch(plt.Circle((0, 0), radius=self.radius, fc=self.color))
        plt.axis('scaled')
        plt.show()  
````

## Creating an instance of a class Circle


Once we have defined our class, we are ready to create **objects** (also called **instances**) from it. To create an object, we simply call the class as if it were a function and pass the required arguments to the constructor.

````python
redCircle = Circle(10, 'red')
````

We can use the `dir` command to get a list of the object's methods. Many of them are default Python methods.

````python
dir(Circle)
````

We can look at the data attributes of the object:

````python
# print the object attribute radius
redCircle.radius
````

````python
# print the object attribute color
redCircle.color
````

We can change the object's data attributes

````python
# Set the object attribute radius

redCircle.radius = 1
redCircle.radius
````

We can draw the object by using the method `drawCircle()`:

````python
# Call the method drawCircle

redCircle.drawCircle()
````

We can increase the radius of the circle by applying the method <code>add_radius()</code>. Let's increases the radius by 2 and then by 5:


````python
# Use method to change the object attribute radius

print('Radius of object:',redCircle.radius)
redCircle.add_radius(2)
print('Radius of object of after applying the method add_radius(2):',redCircle.radius)
redCircle.add_radius(5)
print('Radius of object of after applying the method add_radius(5):',redCircle.radius)
````

Let’s create a blue circle. As the default colour is blue, all we have to do is specify what the radius is:


````python
# Create a blue circle with a given radius

blueCircle = Circle(radius=100)
````

Let's draw this object

````python
# Call the method drawCircle

blueCircle.drawCircle()
````
