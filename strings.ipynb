### Slicing

We can obtain multiple characters from a string using **slicing**

```python
texto = "La otra noche salí a comer tacos." 
```

```python
# printing first 5 characters
texto[0:5]
```

```python
# step through the string two characters at a time
texto[::2]
```

```python
# Step through the string two characters at a time until the 9 character.
texto[0:9:2]
```

### String Multiplication
We can multiply strings by an integer

```python
word = "Git "
3*word
```

### Escape Sequences

They beggins with back slashes `\`. Escape sequences represent strings that may be difficult to input. 

```python
# New line
print("Mi comida favorita\nes la pizza")
```

```python
# Tab
print("Mi comida favorita es\t la pizza")
```

```python
# Put a back slash in a string
print("Mi comida favorita es la \\pizza")
```

### String Manipulation Operations

There are many string operation methods in Python that can be used to manipulate data

`upper()` method converts all characters to uppercase.

```python
text = 'mY keyBOard isn\'T wOrkING.'
new_text = text.upper()
new_text
```

`lower()` method converts all characters to lowercase

```python
text.lower()
```

The method `replace()` replaces the segments of a string. We input the part of the string we want to replace and the string that will replace it.

```python
text = 'git hub git hub git hub git hub'
# replace 'git' by 'nrop'
text.replace('git', 'nrop')
```

You can specify how many occurrences you want to replace with `count`

```python
# Change only the first 2 occurrences
text.replace('git', 'nrop', count = 2)
```

The method `find()` searches for a substring. The argument is the substring to search for and the output is the index where the first occurrence starts. If the method doesn't find anything, it returns `-1`.

```python
text = (
    "Welcome to our website. Here is some random text that does not matter. "
    "Please ignore this section. password: uen1id8*n "
    "More random text follows to simulate a long message."
)

# Find the hidden password
password_first_index = text.find("password")
print(f"The password begins at index {password_first_index}")
```

```python
# An example where it doesn't find anything
text = 'Hello, my name is Clark'
print(text.find('Joseph'))
```

The method `split()` splits the string at the specified separator, and return a list.

**Syntax**

`string.split(sep, maxsplit)`

- `sep` (optional) is a string that specifies the separator. If not provided, the default separator is any whitespace.
- `maxsplit` (optional) is an integer, this specifies the maximum number of splits to perform. If not provided, there is no limit on the number of splits.

```python
name = 'Ana Melchor Segoviano'
name.split()
```

```python
# A comma as separator
ages = '10,45,21,64,13,64,12,93,34'
ages.split(sep = ',')
```

```python
# A comma as separator and split only 3 times
ages = '10,45,21,64,13,64,12,93,34'
ages.split(sep = ',', maxsplit=3)
```
