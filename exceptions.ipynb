## Exception Handling

### Try Except

The `try... except` statement will allow us to execute code that might raise an exception and in the case of any exception or a specific one we can handle or catch the exception and execute specific code. This will allow us to continue the execution of our program even if there is an exception.

Python tries to execute the code in the <code>try</code> block. In this case if there is any exception raised by the code in the <code>try</code> block, it will be caught and the code block in the <code>except</code> block will be executed. After that, the code that comes <em>after</em> the try except will be executed.

```python
# potential code before try catch

try:
    # code to try to execute
except:
    # code to execute if there is an exception
    
# code that will still execute if there is an exception
```

Let's see an example. If a user attempts to divide by zero, the program can continue waiting until the user enters a valid number instead of stopping.

```python
# Program without try... except statement
a = 4
b = int(input('Divide 4 by: '))
print(f'{a/b}')
```

A `try... except` statement can fix this potential problem.

```python
# Program with try... except statement

while(True):
    try:
        a = 4
        b = int(input('Divide 4 by: '))
        print(f'Result = {a/b}')
        break
    except:
        print("Enter a valid number.")
```

The issues could be even greater. The user might also enter invalid input. For instance, they could enter a letter or a symbol.

In this case, the program has another problem: the user might not know why the program doesn't "like" the input. _How can we communicate the exact problem to the user?_

We can catch a specific exception by writing the exception's name after the `except` keyword.

```python
# Program with try... except statement and specific exceptions.

while(True):
    try:
        a = 4
        b = int(input('Divide 4 by: '))
        print(f'Result = {a/b}')
        break
    # handling a dividing by zero error
    except ZeroDivisionError:
        print("You can't divide by zero! Enter a valid number.")
    # Handling an error caused by entering an invalid literal (for example, a letter)
    except ValueError:
        print('You did not provide a number!')
    
```

If an unforeseen problem arises during execution, simply use the `except` keyword on its own (although you should always try to cover all possible specific exceptions).

```python
while(True):
    try:
        a = 4
        b = int(input('Divide 4 by: '))
        print(f'Result = {a/b}')
        break
    # handling a dividing by zero error
    except ZeroDivisionError:
        print("You can't divide by zero! Enter a valid number.")
    # Handling an error caused by entering an invalid literal (for example, a letter)
    except ValueError:
        print('You did not provide a number!')
    # another problems
    except:
        print('Something went wrong!')
    
```

Above, the message "Something went wrong!" is displayed, interrupting the user's input.

### Try Except Else and Finally

<code>else</code> allows one to check if there was no exception when executing the try block. This is useful when we want to execute something only if there were no errors.


<code>finally</code> allows us to always execute something even if there is an exception or not. This is usually used to signify the end of the try except.


```python
# potential code before try catch

try:
    # code to try to execute
except ZeroDivisionError:
    # code to execute if there is a ZeroDivisionError
except NameError:
    # code to execute if there is a NameError
except:
    # code to execute if ther is any exception
else:
    # code to execute if there is no exception
finally:
    # code to execute at the end of the try except no matter what
    
# code that will execute if there is no exception or a one that we are handling
```

Following the examples presented.

```python
while True:
    try:
        a = 4
        b = int(input('Divide 4 by: '))
        print(f'Result = {a / b}')

    # Handling a dividing by zero error
    except ZeroDivisionError:
        print("You can't divide by zero! Enter a valid number.")

    # Handling an error caused by entering an invalid literal (for example, a letter)
    except ValueError:
        print('You did not provide a number!')

    # Handling any other unexpected errors
    except:
        print('Something went wrong!')

    # Runs only if no exception occurred
    else:
        print('The operation completed successfully.')
        break

    # Runs whether or not an exception occurred
    finally:
        print('[Hello from \'finally\' statement. This message always appears.]')
```

### Another useful examples

Handling two errors at a time

```python
# Handling two errors at a time
try:
    a = 4
    b = int(input('Divide 4 by: '))
    print(f'Result = {a / b}')

except (ZeroDivisionError, ValueError):
    print("Error: dividing by zero or invalid input (enter a number)")
```

Catching all exceptions as an object

```python
# Catching an exception as an object
try:
    number = int(input("Enter a number: "))
    print(10 / number)

except Exception as error:
    print(f"An error occurred: {error}")
```

Raising your own exception

```python
# Raising an exception
age = int(input("Enter your age: "))

if age < 0:
    raise ValueError("Age cannot be negative.")

print("Age accepted.")
```
