## What is a Series?

A **Series** is a one-dimensional labeled data structure provided by Pandas. It can store values of any data type, and each value is associated with an **index** that is used to identify and access the data. You can think of a Series as a single column of a table.

```python
import pandas as pd
# Create a Series from a list
data = [10, 20, 30, 40, 50]
s1 = pd.Series(data)
print(s1)
```

### **Index labels**
Each item in a `Series` object has a unique identifier called the index label. By default, it is simply the rank of the item in the `Series` (starting from `0`) but you can also set the index labels manually:

```python
s2 = pd.Series([68, 83, 112, 68], index=["alice", "bob", "charles", "darwin"])
s2
```

You can then use `Series` just like a `dict`:

```python
s2['bob']
```


To make it clear when you are accessing by label or by integer location, it is recommended to **always** use the `loc` attribute when accessing by label, and the `iloc` attribute when accessing by integer location:

```python
# access by integer
s2.iloc[2]
```

```python
# acces by label
s2.loc['charles']
```

Slicing a `Series` also slices the index labels:

```python
s2.iloc[2:]
```

## Useful Series Attributes and Methods

Pandas **Series** provide many built-in **attributes** and **methods** that make data exploration and manipulation much easier.

- **Attributes** store information about the Series, such as its size or index. They are accessed without parentheses (e.g., `series.size`).
- **Methods** perform an action or computation, such as calculating the mean or sorting the data. They are called using parentheses (e.g., `series.mean()`).

Some of the most commonly used attributes and methods are:

- **`values`**: Returns the data as a NumPy array.
- **`index`**: Returns the labels (index) of the Series.
- **`shape`**: Returns a tuple representing the dimensions of the Series.
- **`size`**: Returns the total number of elements.

### Statistical methods

- **`mean()`**: Returns the average of the values.
- **`sum()`**: Returns the sum of all values.
- **`min()`**: Returns the smallest value.
- **`max()`**: Returns the largest value.

### Working with unique values

- **`unique()`**: Returns the distinct values in the Series.
- **`nunique()`**: Returns the number of distinct values.

### Sorting

- **`sort_values()`**: Sorts the Series according to its values.
- **`sort_index()`**: Sorts the Series according to its index labels.

### Handling missing values

- **`isnull()`**: Returns `True` for missing (`NaN`) values and `False` otherwise.
- **`notnull()`**: Returns `True` for non-missing values and `False` otherwise.

- **`apply()`**: Applies a user-defined function to every element of the Series.

## What is a DataFrame?


A **DataFrame** is the primary data structure in Pandas. It represents data in a **two-dimensional table** with labeled rows (index) and labeled columns. Unlike a Series, which stores a single column of data, a DataFrame can contain **multiple columns**, each with its own data type.

DataFrames are widely used for storing, cleaning, analyzing, and manipulating structured data. You can think of a DataFrame as a spreadsheet or a SQL table, but with many powerful tools for data analysis built into Python.

## Creating a DataFrame from a Dictionary

One of the simplest ways to create a DataFrame is from a Python **dictionary**. In this case, each **key** becomes a column name, and each **value** (typically a list or another sequence) becomes the data stored in that column.

All columns should have the same number of elements, since each row represents a single observation.

```python
import pandas as pd

# Create a dictionary
students = {
    "Name": ["Alice", "Bob", "Charlie", "Diana"],
    "Age": [20, 21, 19, 22],
    "Major": ["Physics", "Mathematics", "Computer Science", "Biology"]
}

# Create a DataFrame from the dictionary
df_students = pd.DataFrame(students)

# Display the DataFrame
df_students
```

You can also create a DataFrame by passing a dictionary of `Series` objects:

```python
people_dict = {
    "weight": pd.Series([68, 83, 112], index=["alice", "bob", "charles"]),
    "birthyear": pd.Series([1984, 1985, 1992], index=["bob", "alice", "charles"], name="year"),
    "children": pd.Series([0, 3], index=["charles", "bob"]),
    "hobby": pd.Series(["Biking", "Dancing"], index=["alice", "bob"]),
}
people = pd.DataFrame(people_dict)
people
```

You can access columns pretty much as you would expect. They are returned as `Series` objects:

```python
people['birthyear']
```

## Accessing Data with `loc` and `iloc`

Pandas provides two main ways to access data in a DataFrame:

- **`loc`** selects rows and columns **by their labels** (index and column names).
- **`iloc`** selects rows and columns **by their integer positions**, starting from `0`.

In general:

- Use **`loc`** when you know the **names** of the rows or columns.
- Use **`iloc`** when you know their **positions** in the DataFrame.

The table below summarizes the main differences:

| `loc` | `iloc` |
| :--- | :--- |
| Uses row and column **labels** | Uses integer **positions** |
| Slice endpoints are **included** | Slice endpoints are **excluded** |
| Example: `df.loc["A", "Name"]` | Example: `df.iloc[0, 0]` |

Using `loc`

```python
# Select the row with label "alice"
people.loc["alice"]
```

```python
# Select the value at row "bob" and column "weight"
people.loc["bob", "weight"]
```

```python
# Select multiple rows
people.loc[["alice", "charles"]]
```

```python
# Slice rows by their labels (both endpoints are included)
people.loc["alice":"charles"]
```

Using `iloc`

```python
# Select the first row (position 0)
people.iloc[0]
```

```python
# Select the value at row 1 and column 0
people.iloc[1, 0]
```

```python
# Select specific rows and columns
people.iloc[[0, 1], [0, 1]]
```

```python
# Slice rows and columns by their positions (the endpoint is excluded)
people.iloc[0:3, 2:4]
```

## Accessing Columns with `[]`

A DataFrame column can also be accessed using square brackets `[]`. The argument inside the brackets is the **column label**.

When selecting **one column**, Pandas returns a **Series**. If you instead provide a **list of column names**, Pandas returns a **DataFrame** containing only those columns.

This is one of the most common ways to select columns from a DataFrame.

```python
# DataFrame
people
```

```python
# Select a single column (returns a Series)
people["weight"]
```

```python
# Select multiple columns (returns a DataFrame)
people[["weight", "birthyear"]]
```
